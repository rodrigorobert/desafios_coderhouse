2. Comparar en una misma gráfica las distribuciones de total_bill y tip



```python
#Completar

sns.set(style="darkgrid")

sns.histplot(data=df, x="total_bill", color="skyblue", label="total_bill", kde=True)
sns.histplot(data=df, x="tip", color="red", label="tip", kde=True)

plt.legend() 
plt.show()
```

```python
ig, axs = plt.subplots(2, 2, figsize=(7, 7))

sns.histplot(data=df, x="total_bill", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="tip", kde=True, color="olive", ax=axs[0, 1])

plt.show()
```

# Scatterplot con Plotly

1. Generar un scatter plot donde el eje x es "total_bill" y el eje y es "tip"

```python
#Completar

fig = px.scatter(y=df["tip"], x=df["total_bill"])
fig.show()
```

2. Asignar los colores en base a la variable "day"

```python
#Completar

fig = px.scatter(y=df["tip"], x=df["total_bill"], color=df['day'])
fig.show()
```

3. Modificar el tamaño de los points en base a la variable "size"

```python
#Completar

fig = px.scatter(df, y=df["tip"], x=df["total_bill"], color=df['day'], size='size')
fig.show()
```

4. Generar un scatter plot en 3d donde la variable **"x"** es "total_bill", la variable **"y"** es "tip" y la variable **"z"** es "size"

```python
#Completar

fig = px.scatter_3d(df, y="tip", x="total_bill", z='size')
fig.show()
```

5. Modificar el grafico anterior y asignar un color en base a la variable "day"


```python
#Completar

fig = px.scatter_3d(df, y="tip", x="total_bill", z='size', color='day')
fig.show()
```
