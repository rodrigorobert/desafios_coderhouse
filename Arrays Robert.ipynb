**2. Recordar los ejercicios con funciones para cálculo de factorial y suma de serie de la Clase 02. Repetir ambos ejercicios, pero ahora utilizar las nuevas operaciones aprendidas con los ndarrays**

```python
# En esta celda creamos la función factorial con numpy:

def factorial(y):
    factorial = np.arange(2, y + 1).prod()
    return factorial
```

```python
# En esta celda la ponemos a funcionar:

factorial(5)
```

```python
# si usamos la librería:

np.math.factorial(5)
```

```python
# En esta celda creamos la función para sumer series con numpy:

def sum_serie(n1,n2):
  z = np.arange(n1, n2+1)
  suma = np.sum(z)
  return suma
```

```python
# En esta celda la ponemos a funcionar:

sum_serie(10,11)
```

Data cleaning

```python
# Como primer filtro eliminamos todos los datos que no son de Argentina

df_arg = df[df['Estoy trabajando en'] == 'Argentina']
```

```python
#Visializamos cantidad de filas y columnas para ver cómo se va modificando el data set

print(df_arg.shape)
```

```python
# Visualizamos los nombres de las columnas para identificar aquellas sobre las qeu queremos trabajar

columns_names = df_arg.columns.values
print(columns_names)
```

```python
# Eliminamos de la column que nos interesa todas las filas que son NaN

df_arg = df_arg.dropna(subset=['Salario mensual NETO (en tu moneda local)'])
```

```python
# Convertimos todos los datos de la columna que nos interesa a float, sin esto después las operaciones en numpy nos dan error

df_arg['Salario mensual NETO (en tu moneda local)'] = df_arg['Salario mensual NETO (en tu moneda local)'].astype('float')
```

```python
# Convertimos el pandas dataset a un array de numpy

arr = df_arg.to_numpy()
arr.shape
```

```python
# Hacemos un slicing -corte- de la columa que nos interesa

neto = arr[:,48:49]
```

PROCESAMIENTO DE DATOS

```python
# Visualizamos como hemos pasado de 6415 filas y 81 columnas a 5883 filas y una columna en forma de array
neto.shape
```

```python
# Verificamos que tengamos los datos númericos que necesitamos

print(neto)
```

```python
# Calculamos promedio

np.mean(neto)
```

```python
# Calculamos mediana

np.median(neto)
```

# Análisis comparativo

Considerando que media y mediana no coinciden podemos decir que la distribución de los sueldos netos no es normal. Dado que la media es menor que la mediana la distribución está sesgada a la izquierda, es decir que proporcionalmente son muchos los que cobran poco y pocos los que cobran mucho.
