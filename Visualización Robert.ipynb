**2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?**

```python
salarios = df['Salary']
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax.hist(salarios.values.flatten(), bins=70)
ax.set_title('Histograma de salarios')
ax.set_xlabel('Intervalos de salarios (usd)')
ax.set_ylabel('Frecuencia absoluta')
```

Por los se ve en el gráfico los salarios más populares estan entre 50000 y 100000 dolares al año, 60 o 65K deben ser la columna más alta.

```python
# Aunque la MODA no es una medida recomendable para estadistica, si calculamos la moda vemos que se acerca a los valores del cuadro anterior. 
sal_moda = pd.Series(df[['Salary']].values.flatten()).mode()[0]
print(f'El salario más frecuente es de {sal_moda}')
```

```python
# Si calculamos la media sucede lo mismo, se acerca al valor antes mencionado:

sal_media = salarios.median()
print(f'La media del salario es de {sal_media}')
```

**3. Realizar un gráfico de violín con los salarios, discriminados por género.**

**4. Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire) - numero de despidos vs la fecha de contratacion**

```python
# Construimos la variable temporal

date = pd.to_datetime(df['DateofHire'])
```

```python
# Ralizamos el cuadro con el Id de cada empleado porque es un número que no se repite y eso nos permitira ver la desidad de puntos en cada año
x = date
y = df['EmpID']

plt.plot_date(x, y)
plt.gcf().set_size_inches(8, 6)
plt.tight_layout()
plt.show()
```

```python
# Aquí graficamos los mismo pero con barras para visualizar mejor la proporción, las barras coinciden con los años donde había más densidad de puntos.

ig, ax = plt.subplots(figsize=(8, 6))
ax.hist(x.values.flatten(), bins=70, density=True)
ax.set_title('Cantidad de despidos en el tiempo')
ax.set_xlabel('Años')
ax.set_ylabel('Proporción de empleados despedidos')
```

EL mayor número de despidos fue entre los años 2011 y 2015
