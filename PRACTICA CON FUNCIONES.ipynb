**1. Escribir una función para calcular el factorial de un número cualquiera.**

```python
# En esta celda definimos la función:

def factorial(x): 
    if x < 0: 
        print("No se puede calcular factorial de números menores a 0")

    elif x == 0: 
        return 1
        
    else: 
        y = 1
        while(x > 1): 
            y *= x 
            x -= 1
        return y 
```

```python
# En esta celda la ponemos a funcionar:

factorial(0)
```

**2. Escribir una función para calcular la suma de una serie comenzando por un número cualquiera y terminando en otro número que debe ser mayor al primero**

```python
# En esta celda definimos la función:

def sum_serie(n1,n2):
  if n2 <= n1:
    print('error')
  else:
    for i in range(n1,n2):
      n1 += i+1
    print(n1)
```

```python
# En esta celda la ponemos a funcionar:

sum_serie(10,11)
```
