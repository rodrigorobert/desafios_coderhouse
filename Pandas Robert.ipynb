**2. Separar nombres de apellidos en dos columnas distintas usando la coma como separador**

```python
# Separamos la columnas en nombre y apellido

split_name = df["Athlete"].str.split(',', expand=True)
split_name.drop([2], axis=1)
```

```python
# Nombramos las nuevas columnas

split_name.columns = ['last_name', 'first_name', 'index']
```

```python
# Las unimos al df original, borramos la columna anterior y las ordenamos

df_split = pd.concat([df, split_name], axis=1)
df_split.drop(['index', 'Athlete'], axis=1)
df_ordenado = df_split.reindex(columns=['City',	'Year',	'Sport',	'Discipline',	'Event',	'last_name',	'first_name',	'Gender',	'Country_Code',	'Country',	'Event_gender',	'Medal'])
```

```python
# Comprobamos la operacion

df_ordenado.head()
```

**3. Obtener medidas resumen del conjunto de datos. Cuál es el país que ganó más medallas?**

```python
medals = df.groupby(['Country']).count()
medals = medals.drop(['City',	'Year',	'Sport',	'Discipline',	'Event',	'Athlete', 'Gender',	'Country_Code',	'Event_gender'], axis=1)
medals
```

```python
m = medals["Medal"].max(skipna=True)
c = medals["Medal"].idxmax(skipna=True)

print(f'El país con más medallas es {c} con {m} medallas')
```

```python
prom = int(medals.mean(axis='rows', skipna=True))
print(f'El promedio de medallas por país es de {prom}')
```

```python
media = int(medals.median(axis='rows', skipna=True))
print(f'El media de medallas por país es de {media}')
```

```python
# Se puede comprobar que coinciden los datos anteriores con los mencionados más arriba

medals.describe()
```
